```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
import re

def text_to_list(text):
    # Split the text based on spaces and punctuation
    return re.findall(r'\w+', text)
```

```python
md['genres'] = md['genres'].fillna('[]')
md['genres'] = md['genres'].apply(convert)
md['overview'] = md['overview'].fillna('[]')
md['overview'] = md['overview'].apply(text_to_list)
```

```python
def convert_string_to_list(input_data):
    if isinstance(input_data, str):

        cleaned_string = input_data.lower().replace(" ", "")

        return [cleaned_string]
    elif isinstance(input_data, list):
        result_list = []
        for item in input_data:
            if isinstance(item, str):
                cleaned_string = item.lower().replace(" ", "")
                result_list.append(cleaned_string)
            else:
                print(f"Warning: Non-string item '{item}' ignored.")
        return result_list
    else:
        print("Error: Input must be a string or a list of strings.")
        return []

def combine_strings(input_list):
    if isinstance(input_list, list):

        combined_string = ''.join(input_list).lower()

        return [combined_string]
    else:
        print("Error: Input must be a list of strings.")
        return []
```

```python
md['original_title'] = md['original_title'].fillna('[]')
md['original_title'] = md['original_title'].apply(convert_string_to_list)
md['original_title'] = md['original_title'].apply(combine_strings)
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
def convert_int(x):
    try:
        return int(x)
    except:
        return np.nan
```

```python
links = pd.read_csv('links.csv')
links = links[links['tmdbId'].notnull()]['tmdbId'].astype('int')

# final_df = final_df.drop([19730, 29503, 35587])


final_df['id'] = final_df['id'].astype('int')

smd = final_df[final_df['id'].isin(links)]
smd.shape
smd.head()
```

```python
id_map = pd.read_csv('links.csv')[['movieId', 'tmdbId']]
id_map['tmdbId'] = id_map['tmdbId'].apply(convert_int)
id_map.columns = ['movieId', 'id']
id_map = id_map.merge(smd[['title', 'id']], on='id').set_index('title')
```

```python
id_map.head()
```

```python
smd = smd.drop_duplicates(subset='title', keep='first')
indices = pd.Series(smd.index, index=smd['title'])
indices_map = id_map.set_index('id')
```

```python
def hybrid(userId, title):

    if title not in indices:
        raise ValueError(f"Title '{title}' not found in the dataset.")


    idx = indices[title]
    tmdbId = id_map.loc[title]['id'].iloc[0] if isinstance(id_map.loc[title]['id'], pd.Series) else id_map.loc[title]['id']
    movie_id = id_map.loc[title]['movieId'].iloc[0] if isinstance(id_map.loc[title]['movieId'], pd.Series) else id_map.loc[title]['movieId']


    sim_scores = list(enumerate(similarity[int(idx)]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]


    movie_indices = [i[0] for i in sim_scores]

    valid_indices = [i for i in movie_indices if i < len(smd)]



    movies = smd.iloc[valid_indices][['title', 'id']]


    movies['est'] = movies['id'].apply(
        lambda x: svd.predict(userId, indices_map.loc[x]['movieId'].iloc[0]).est
        if isinstance(indices_map.loc[x]['movieId'], pd.Series)
        else svd.predict(userId, indices_map.loc[x]['movieId']).est
    )


    movies = movies.sort_values('est', ascending=False)
    return movies.head(3)
```

```python
hybrid(1, 'Batman')
```

```python
def final(user,title):
    print(recommend(title))
    print(hybrid(user, title))
```
